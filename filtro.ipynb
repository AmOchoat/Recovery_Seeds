```python
import pathlib
```

```python
# Path es resultados_ml_v2
Path = pathlib.Path("resultados_ml_v2")
# Listar todos los archivos en el directorio y subdirectorios
all_files = list(Path.rglob("*"))
```

```python
all_files
```

```python
"""
Abre cada uno de los archivos y lee su contenido
Algunos archivos tienen esta estructura:

{
  "status": "EMPTY",
  "group_id": "W192_mu4_w32_eta2_a0.001_b0.2_BT2p30_BM2p30",
  "error": "",
  "W": 192,
  "mu": 4,
  "w": 32,
  "eta": 2,
  "alpha": 0.001,
  "beta": 0.2,
  "Btime": 1073741824,
  "Bmemory": 1073741824,
  "timestamp": 1762311290.540797
}


Quiero que los que tengan estatus empty los mueva a una carpeta llamada vacios
"""
for file_path in all_files:
    # skip if not a regular file
    if not file_path.is_file():
        continue
    # Open and read, then close before trying to move the file
    with open(file_path, 'r', encoding='cp1252', errors='ignore') as file:
        content = file.read()
    if '"status": "EMPTY"' in content:
        # Mover el archivo a la carpeta 'vacios'
        vacios_dir = Path / "vacios"
        # ensure parent directories and avoid exception if already exists
        vacios_dir.mkdir(parents=True, exist_ok=True)
        new_path = vacios_dir / file_path.name
        # If target already exists, skip or create a unique filename.
        if new_path.exists():
            # Skip to avoid FileExistsError; alternatively you could generate a unique name
            print(f"Skipping move for {file_path.name}  target already exists at {new_path}")
            continue
        try:
            file_path.rename(new_path)
        except PermissionError:
            # If rename fails (file locked), skip or handle as needed
            print(f"Could not move {file_path}  file is in use.")
```
